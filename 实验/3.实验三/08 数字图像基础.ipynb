## 实验一

```python
# 导入必要模块
import matplotlib.pyplot as plt # type: ignore
from PIL import Image

# 解决中文
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

img = Image.open("lena_std.tif")
img.thumbnail((256,256))
# RGB 三通道分离
img_r,img_g,img_b = img.split()

plt.figure(figsize=(7,7))

img_r_small = img_r.resize((50,50))   # 缩放图片
plt.subplot(221)
plt.axis("off")
plt.imshow(img_r_small, cmap = "gray")
plt.title("R-缩放", fontsize=14)

img_g_trans = img_g.transpose(Image.FLIP_LEFT_RIGHT)\
                    .transpose(Image.ROTATE_270)
plt.subplot(222)
plt.imshow(img_g_trans, cmap = "gray")
plt.title("G-镜像+旋转", fontsize=14)

img_b_region = img_b.crop((0,0,150,150))
plt.subplot(223)
plt.axis("off")
plt.imshow(img_b_region, cmap = "gray")
plt.title("B-裁剪", fontsize=14)

# img_con = [img_r, img_g, img_b]
plt.subplot(224)
plt.axis("off")
plt.imshow(img)
plt.title("RGB", fontsize=14)

plt.suptitle("图像基本操作",fontsize=20,color="blue")

plt.subplots_adjust(wspace=0.6, hspace=0)  # 增大水平间距 减小标题和子图之间的间隙
```

```python
import numpy as np
import random
import matplotlib.pyplot as plt
from keras.datasets import mnist

(train_x, train_y), (test_x, test_y) = mnist.load_data()

# 测试集和数据集的长度
print("train_x set:",len(train_x))
print("train_y set:",len(train_y))
print("test_x set:",len(test_x))
print("test_y set:",len(test_y))

# 图像数据和标记数据的形状
print("train_x:",train_x.shape, train_x.dtype)
print("train_y:",train_y.shape, train_y.dtype)
print("test_x:",test_x.shape, test_x.dtype)
print("test_y:",test_y.shape, test_y.dtype)

plt.figure(figsize=(10,10))
plt.suptitle("MNIST测试集样本",fontsize=20,color="red")

for i in range(16):
    plt.subplot(4,4,i+1)
    plt.axis("off")
    rd = random.randint(0,10000-1)
    plt.imshow(test_x[rd], cmap="gray")
    plt.title(f"标签值：{test_y[rd]}", fontsize=14)


plt.figure(figsize=(10,5))

plt.subplot(121)
plt.imshow(test_x[16], cmap="gray")
plt.title("原图", fontsize=14)

plt.subplot(122)
test_regin = test_x[16][:14,14:28]
plt.imshow(test_regin, cmap="gray")
plt.title("右上角1/4", fontsize=14)



def solve(img, stepx1, stepy1, title1, stepx2, stepy2, title2):
    plt.figure(figsize=(10,3))

    plt.subplot(131)
    plt.imshow(img, cmap="gray")
    plt.title("原图", fontsize=14)

    plt.subplot(132)
    test_sub = img[::stepx1,::stepy1]
    plt.imshow(test_sub, cmap="gray")
    plt.title(title1, fontsize=14)

    plt.subplot(133)
    test_sub = img[::stepx2,::stepy2]
    plt.imshow(test_sub, cmap="gray")
    plt.title(title2, fontsize=14)

solve(test_x[0],2,None,"隔2行采样",4,None,"隔4行采样")

solve(test_x[16],2,None,"隔2行采样",4,None,"隔4行采样")

solve(test_x[32],2,2,"隔2行2列采样",4,4,"隔4行4列采样")

solve(test_x[48],2,2,"隔2行2列采样",4,4,"隔4行4列采样")
```
