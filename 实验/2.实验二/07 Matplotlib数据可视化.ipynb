## 题目一：
请下载波士顿房价数据集，并按照以下要求访问和可视化数据。

(1) 下载波士顿房价数据集，并将所有训练集数据读取到NumPy数组x和y中。其中，x为属性数组，y为标记数组。

(2) 使用所有训练集数据，可视化为如图1所示形式。

(3) 展示所有属性并提示用户输入属性序号，如图2所示；根据用户选择的属性输出对应的散点图，如图3所示。

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_boston

# 解决中文
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

# (1) 下载波士顿房价数据集，并将所有训练集数据读取到NumPy数组x和y中
boston = load_boston()
x = boston.data
y = boston.target

# 创建包含每个属性与房价之间关系的散点图
plt.figure(figsize=(15, 15))
for i, feature_name in enumerate(boston.feature_names, start=0):
    plt.subplot(4, 4, i+1)  # 划分 4x4的网格
    plt.scatter(x[:, i], y)
    plt.xlabel(feature_name)
    plt.ylabel("Price($1000\\'s)")
plt.tight_layout()
plt.suptitle("各个属性与房价的关系", x=0.5, y=1.02, fontsize=20)
plt.show()

# (3) 展示所有属性并提示用户输入属性序号
print("属性序号:")
for i, feature_name in enumerate(boston.feature_names):
    print(f"{i+1} -- {feature_name}")

# 提示用户输入属性序号
selected_feature_index = int(input("请输入属性序号: "))
selected_feature_index -= 1
# 根据用户选择的属性输出对应的散点图
plt.figure(figsize=(4, 4))
plt.scatter(x[:, selected_feature_index], y, marker='o', c='blue', alpha=0.7)
plt.title(f'{selected_feature_index+1}.{boston.feature_names[selected_feature_index]} -Price')
plt.xlabel(boston.feature_names[selected_feature_index])
plt.ylabel("Price($1000\\'s'")
plt.grid(True)
plt.show()
```

## 题目二：
请下载鸢尾花数据集，并参照图4进行数据可视化。

⑴  两两组合所有属性，并以散点图形式可视化；

⑵ 对角线为各个属性的直方图，其数值为该属性在不同取值下的样本数。

提示：绘制直方图函数 plt.hist(x,, color, edgecolor)

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# 加载鸢尾花数据集
iris = load_iris()
data = iris.data
target = iris.target
feature_names = iris.feature_names

fig = plt.figure('Iris Data', figsize=(15, 15))
fig.suptitle('Iris', fontsize=16)
for i in range(4):
    for j in range(4):
        if i == j:
            plt.subplot(4, 4, 4 * i + (j + 1))
            plt.hist(data[:, i], align='mid', color='skyblue', edgecolor='black')
            plt.ylabel('Count')
        else:
            plt.subplot(4, 4, 4 * i + (j + 1))
            plt.scatter(data[:, j], data[:, i], c=target[:], cmap='brg')
        if i == 0:
            plt.title(feature_names[j])
        if j == 0:
            plt.ylabel(feature_names[i])
            
plt.show()
```

## 题目三(附加题)：
  表1是我国2007-2022年的国内生产总值（GDP）统计表，请使用Matplotlib绘制折线图。

题目要求：

(1) 标题为“2007-2022年我国国内生产总值统计图”，字体颜色为黑色，大小为16；

(2) 蓝色图线是全年GDP（亿元），线条宽度为2.5；

```python
import matplotlib.pyplot as plt

# 年份
years = list(range(2007, 2023))

# 我国2007-2022年的国内生产总值（GDP）数据（假设这是实际数据）
gdp = [270092, 319244, 348517, 412119, 487940, 538580, 592963, 
       643563, 688858, 746395, 832035, 919281, 986515, 1010000, 
       1140000, 1210000]

# (1) 绘制折线图
plt.figure(figsize=(10, 6))
plt.plot(years, gdp, color='blue', linewidth=2.5)
plt.title("2007-2022年我国国内生产总值统计图", color='black', fontsize=16)
plt.xlabel("年份")
plt.ylabel("全年GDP（亿元）")
plt.grid(True)
plt.show()
```
